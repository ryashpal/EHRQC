```python
import os
import pandas as pd
```

```python
# path = os.getcwd() + '/../../data/mimic_top_numeric_vitals.csv'
path = os.getcwd() + '/../../data/mimic_top_labs.csv'
dataDf = pd.read_csv(path)
```

```python
dataDf.head()
```

```python
dataDf = dataDf.iloc[: , 1:]
```

```python
dataDf.set_index('hadm_id', inplace=True)
```

```python
dataDf.head()
```

```python
from nipals import nipals

nip = nipals.Nipals(dataDf)
nip.fit(ncomp=2, maxiter=1000, center=True, scale=True, tol=0.1)
```

```python
dataDf[:1000]
```

```python
from statsmodels.multivariate.pca import PCA

pc = PCA(data=dataDf[:10000], ncomp=1, missing='fill-em')
```

```python
imputedDf = pd.DataFrame(pc._adjusted_data)
imputedDf
```
