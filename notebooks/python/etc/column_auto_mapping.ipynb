```python
from fuzzywuzzy import fuzz

for column_name in column_names_list:
    max_score = 0
    for plotting_column_name in plotting_column_names_list1:
        if column_name and plotting_column_name:
            score = fuzz.token_sort_ratio(column_name, plotting_column_name)
            if score > max_score:
                max_score = score
                best_matching_column = plotting_column_name
    print(column_name + ' <=> ' + best_matching_column + ' = ' + str(max_score))
```
